## Missing Value Imputation 

```python
isnull_sum=X_train.isnull().sum()
```

```python
isnull_sum
```

```python
num_var=X_train.select_dtypes(include=["int64","float64"]).columns
```

```python
num_var
```

```python
num_var_miss=[]
for i in num_var:
    if X_train[i].isnull().sum()>0:
        num_var_miss.append(i)
        
```

```python
num_var_miss
```

```python
cat_var=X_train.select_dtypes(include=["object"]).columns
```

```python
cat_var
```

```python
cat_var_miss=[]
for i in cat_var:
    if X_train[i].isnull().sum()>0:
        cat_var_miss.append(i)
```

```python
cat_var_miss
```

```python
num_var_mean=['LotFrontage']
num_var_median=['MasVnrArea', 'GarageYrBlt']
cat_var_const=['Alley','MasVnrType','BsmtQual','BsmtCond']
cat_var_mode=['Electrical','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical']
```

```python
num_var_mean_imputer = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean'))])
num_var_median_imputer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median'))])
cat_var_constant = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing'))])
cat_var_mode_imputer = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent'))])
```

```python
preprocessor = ColumnTransformer(transformers=[('mean_imputer', num_var_mean_imputer, num_var_mean),
                               ('median_imputer', num_var_median_imputer, num_var_median),
                               ('const_imputer', cat_var_constant, cat_var_const),
                               ('mode_imputer', cat_var_mode_imputer, cat_var_mode)])
```

```python
preprocessor.transform
```

```python
preprocessor.fit(X_train)
```

```python
preprocessor.named_transformers_['const_imputer'].named_steps['imputer'].statistics_
```
